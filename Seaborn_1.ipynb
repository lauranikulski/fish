#### Definition of terms used 

##### Demersal 
Species of demersal fish inhabit the bottom of the ocean. Key demersal species fished by the UK fleet include cod, haddock and whiting. 

##### Pelagic
Pelagic fish inhabit the water column (not near the sea bed or shore). The two main pelagic species fished by the UK fleet are mackerel and herring.  

##### Shellfish
Shellfish include various species of molluscs (e.g. scallops, whelks) and crustaceans (e.g. crabs and nephrops).

##### Value
In this publication the value in thousand of pounds (£000's) is reported. This is the value fishers received for their landings at first sale as recorded on sales notes from Registered Buyers and Sellers of fish.

##### Quantity
The quantity in tonnes in reported. This is the live weight of fish caught and landed by fishers.

### Fuffing around with Dataframes

#### Weight 

```python
sns.scatterplot(data=df_weight, x="species", y="weight_t", hue="species") # Scatterplot
```

#### Value

```python
sns.catplot(x="species", y="price_t", data=df_value, kind='swarm') # Scatterplot
```

#### Weight 

```python
sns.relplot(
    data=df_weight,
    x="species", y="weight_t", col="species",
    hue="species", style="species"
)
# statistical relationships plot 
```

#### Value 

```python
sns.relplot(
    data=df_value,
    x="species", y="price_t", col="species",
    hue="species", style="species"
)
```

```python
sns.boxenplot(x="species", y="weight_t",
              color="b",
              scale="linear", data=df_weight)
```

## Using Multivariate Graphs / Multiple Views on Data

```python
df = pd.read_csv('../fish/multivariate.csv')
```

```python
g = sns.lmplot(
    data=df,
    x="weight_t", y="price_t", hue="species",
    height=5
)
# regression model
```

```python
sns.jointplot(data=df, x="weight_t", y="price_t", hue="species", kind="hist")
# interface to a class called JointGrid
```

```python
sns.kdeplot(data=df, x="weight_t", y='price_t', hue="species", multiple="stack")
# kernel density estimate
```

```python
df = pd.read_csv('../fish/multivariate.csv')
sns.set_theme(style="ticks")

g = sns.jointplot(
    data=df,
    x="weight_t", y="price_t", hue="species",
    kind="kde",
)
```

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, x="weight_t", y="price_t", hue="species", ax=axs[0])
sns.histplot(data=df, x="weight_t", hue="price_t", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```
